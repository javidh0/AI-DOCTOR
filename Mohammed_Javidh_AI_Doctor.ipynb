```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("/content/Training.csv")
df_train.head()
```

```python
X = df_train.iloc[:, :-2].values
y = df_train.iloc[:, -2].values
```

```python
print(X)
print(y)
```

```python
dic = dict()
y_t = list()
cnt = 0
for i in range(y.shape[0]):
  if(y[i] not in dic):
    dic[y[i]] = cnt
    cnt+=1
  y_t.append(dic[y[i]])
```

```python
from sklearn.neighbors import KNeighborsClassifier
cls = KNeighborsClassifier(n_neighbors = 5)
cls.fit(X, y_t)
```

```python
df_test = pd.read_csv("/content/Testing.csv")
df_test.head()
```

```python
X_test = df_test.iloc[:, :-1].values
y_test = df_test.iloc[:, -1].values
```

```python
y_test_t = []
for i in range(len(y_test)):
  y_test_t.append(dic[y_test[i]])
print(y_test_t)
```

```python
y_pred = cls.predict(X_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred, y_test_t))
```
